```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
data = pd.read_excel('유동인구데이터처리/PRE_SNS_date_내용X.xlsx', encoding='utf-8')
```

```python
data['CNT']=1
```

```python
del data['TITLE']
```

```python
data
```

```python
data_CNT=data.groupby(['Year','Month','Day','Hour'])['CNT'].sum().reset_index().astype('int')
```

```python
data_CNT.to_csv('유동인구데이터처리/PRE_SNS_CNT_제거X.csv')
```

```python
data.head()
```

```python
#DATE에 값을 일까지만 나타내기
data['DATE']=[str(yyyymmddhh)[:8] for yyyymmddhh in data['DATE']]
```

```python
data.head()
```

```python
#시간을 Datetime으로 바꾸기
data['Datetime'] = data['DATE'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
#정렬 및 Datetime 인덱스화
data=data.sort_values('Datetime')
data=data.set_index('Datetime')
```

```python
#일별로 SNS count 값을 구한다.
df_day=data.resample('D',how={'CNT':np.sum})
df_day_conut=data.resample('D',how={'CNT':np.sum})
```

```python
# 저장
df_day.to_csv('유동인구데이터처리/PRE_day_SNS_CNT.csv')
```

```python
#월별로 SNS count 값을 구한다.
df_month=data.resample('M',how={'CNT':np.sum})
```

```python
#주별로 SNS count 값을 구한다.
df_week=data.resample('W',how={'CNT':np.sum})
```

```python
#평일로 SNS count 값을 구한다.
df_weekday=data.resample('B',how={'CNT':np.sum})
```

```python
# 일별의 데이터값을 정규화 시키기 (그래프로 비교 분석 위해) -> min_max scaling 이용
from sklearn.preprocessing import MinMaxScaler
Min_Max_Scaler=MinMaxScaler()
fitted=Min_Max_Scaler.fit(df_day)
df_day2=Min_Max_Scaler.transform(df_day)
df_day['CNT']=df_day2
```

```python
df_day.head()
```

```python
#지역, 일별 그래프로 SNS 카운트 비율 표현하기
num=0
my_dpi=96
plt.figure(figsize=(36,36), dpi=my_dpi)
plt.style.use('seaborn-darkgrid')
plt.subplot(2,2,1)
plt.scatter(df_day_conut.index, df_day_conut['CNT'].values,color='g', label='Count')
plt.title('SNS-Day Count')
plt.subplot(2,2,2)
plt.plot(df_month.index, df_month['CNT'].values,color='g', label='Count')
plt.title('SNS-Month Count')
plt.subplot(2,2,3)
plt.plot(df_week.index, df_week['CNT'].values,color='g', label='Count')
plt.title('SNS-Week Count')
plt.subplot(2,2,4)
plt.plot(df_weekday.index, df_weekday['CNT'].values,color='g', label='Count')
plt.title('SNS-Week Count')
#plt.text(0.5, 0.02, '월', ha='center', va='center')
#plt.text(0.06, 3, '각각의 비율', ha='center', va='center', rotation='vertical')

plt.savefig('지역,요일별유동인구/SNS,일별,월별,주별,평일별 count.png', dpi=my_dpi*2, bbox_inches='tight', pad_inches=0.0)
```
