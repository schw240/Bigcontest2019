sns 데이터 전처리
-------------------------------------------------------------------------------
> 1. 각 채널(blog, news, sns)별로 해당일자의 평균적인 sns 건수를 구해서 더함
> 2. 큰 값이므로 log값을 취함

```python
rc('font', family='Century Gothic')
```

```python
sns = pd.read_csv('./Pre_sns_data.csv', encoding='utf-8')
```

```python
from datetime import datetime, timedelta

sns['DATE']=pd.to_datetime(sns['DATE'], format='%Y-%m-%d')
```

```python
sns['DATE']
```

```python
a = []
b = []
c = []

for i in sns['DATE']:
    y = i.year
    m = i.month
    d = i.day
    a.append(y)
    b.append(m)
    c.append(d)

sns['YEAR'] = a
sns['MONTH'] = b
sns['DAY'] = c
    
    
```

```python
sns.head()
```

```python
sns_count = sns.groupby(['YEAR','MONTH','DAY','SECTION'])['Count'].agg(np.sum)
```

```python
sns_count = pd.DataFrame(sns_count)
```

```python
sns_count = sns_count.groupby(['YEAR','MONTH','DAY'])['Count'].sum()
```

```python
sns_count = pd.DataFrame(sns_count)
```

```python
#큰수는 데이터에 로그값을 취합니다.
sns_count['Count'] =np.log(sns_count['Count'])
```
