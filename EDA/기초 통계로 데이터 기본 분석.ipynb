```python
### 범주형 변수 파악
# SNS 만 존재함
cat_cols=[col for col in data_sns.columns[:] if data_sns[col].dtype == 'object']
data_sns[cat_cols].describe()
# 블로그가 가장 빈도가 높다. TITLE에서는 미세먼지가 가장 빈도가 높다
# 총 블로그가 302903개를 차지하고 있으며 TITLE에서 미세먼지가 들어가는 빈도는 5188개 이다. -> 추후 contents 부분도 확인해보자.
#  (여기서 '미세먼지' 라고만 들어간 것이 5188개 있다는 의미이다.)
# SECTION에는 총 3개의 분류로 나누어져 있다. (수치형 변수로 바꿔줄 필요가 있다.)
# 총 777832개의 데이터 개수가 있지만 현재 TITLE에는 빈 데이터인 133개(777832-777699)가 존재한다.
```

```python
# 범주형 변수의 고유값 출력
for col in cat_cols:
    uni=np.unique(data_sns[col].astype('str'))
    print('\n')
    print('#col= {}, 개수={}, unique={}'.format(col, len(uni), uni))
```
