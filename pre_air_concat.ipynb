데이터 불러오기
-------------------------------------------------------------------------------
>  노원구/종로구로 저장된 데이터를 하나의 파일로 통합하기 위하여 데이터를 불러와 concat하여줍니다.


```python
pre_nowon = pd.read_csv('pre_nowon.csv')
pre_jongro = pd.read_csv('pre_jongro.csv')
```

```python
pre_air = pd.concat([pre_nowon, pre_jongro])
```

```python
pre_air['Datetime'] = pre_air['Year'].astype('str') + pre_air['Month'].astype('str') + pre_air['Days'].astype('str')
f1 = lambda x: '%s/%02d/%02d'%(x[1], x[2], x[3])
pre_air['Datetime'] = pre_air['Datetime'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))


```

```python
pre_air
pre_air.to_csv('pre_air.csv', index = False)
```

아웃 라이어 제거
-------------------------------------------------------------------------------
>  환경기상데이터의 아웃라이어를 제거하기 위하여 'IsolationForest'를 사용합니다.


```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import IsolationForest

clf = IsolationForest(random_state=1)
clf.fit(pre_air[['pm25','pm10']])
y_pred_outliers = clf.predict(pre_air[['pm25','pm10']])
```

```python
out = pd.DataFrame(y_pred_outliers)
out = out.rename(columns={0: "out"})
```

```python
out
```

```python
pre_air = pre_air.reset_index(drop=True)
out = out.reset_index(drop=True)
pre_air_out = pd.merge(left=pre_air, right=out,left_index=True, right_index=True)
```

```python
pre_air_out
```

```python

pre_air_out = pre_air_out[pre_air_out.out != -1]
#race_an1.out
pre_air_out = pre_air_out.drop(['out'],axis=1)

pre_air_out.to_csv('./pre_air_out.csv',index=False)
print(pre_air_out.shape)
```
