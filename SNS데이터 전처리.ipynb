```python
# 데이터 합침
data_sum=pd.concat([data,data2,data3,data4,data5,data6,data7,data8])
```

```python
data_sum=data_sum.reset_index()
```

```python
del data_sum['index']
```

```python
#날짜를 연, 월, 일, 시, 분만 해서 다른 열에 넣어 놓기
day=[int(str(yyyymmddhhmmss)[:4]) for yyyymmddhhmmss in data_sum['DATE']]
day_2=[int(str(yyyymmddhhmmss)[4:6]) for yyyymmddhhmmss in data_sum['DATE']]
day_3=[int(str(yyyymmddhhmmss)[6:8]) for yyyymmddhhmmss in data_sum['DATE']]
day_4=[int(str(yyyymmddhhmmss)[8:10]) for yyyymmddhhmmss in data_sum['DATE']]
day_5=[int(str(yyyymmddhhmmss)[10:12]) for yyyymmddhhmmss in data_sum['DATE']]
data_sum['Year']=day
data_sum['Month']=day_2
data_sum['Day']=day_3
data_sum['Hour']=day_4
data_sum['Minute']=day_5
```

```python
#데이터를 연도, 월, 일, 시간으로 정렬하기
data_sort=data_sum.sort_values(['Year','Month','Day','Hour','Minute'])
```

```python
data_sort=data_sort.reset_index()
```

```python
#DATE 열에 분, 초를 지움
day_6=[int(str(yyyymmddhhmmss)[:10]) for yyyymmddhhmmss in data_sort['DATE']]
data_sort['DATE']=day_6
```

```python
# 새로운 열에 날짜에 맞춰서 요일 추가하기
day=[6,0,1,2,3,4,5]
i=0
day_sum=[]
for (n1,n2,n3), group in data_sort.groupby(['Year','Month','Day']):
    group['date']=day[i]
    day_sum.append([group['date']])  # date라는 새로운 열에 값을 대입
    if i==6:
        i=0
    else:
        i+=1
```

```python
# 요일 data_sort에 집어넣기
data_sort=data_sort.reset_index()
list_=[]
for i in day_sum:
    list_.append(i[0])
list_1=[]
for i in list_:
    for j in i:
        list_1.append(j)
data_sort['Week_day']=list_1
```

```python
#date를 나눈 파일을 저장
data_sort.to_excel('유동인구데이터처리/PRE_SNS_date_내용X.xlsx', encoding='utf-8',index=False)
```
