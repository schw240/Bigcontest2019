```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
import re
```

```python
#데이터를 클렌징 하기
def cleansing(sentence):
    return re.sub('[^가-힝]',' ', sentence)
```

```python
# 형태소로 나누기 위한 함수
def get_splited(path):
    analyzer = Okt()
    output = analyzer.morphs(cleansing(path))
    return output
```

```python
#문서 내에서 한 단어의 등장 횟수를 출력하는 함수
def tf(doc, word):
    return doc.count(word)
```

```python
### 형태소로 본문 내용을 나눈 후에 '미세먼지' 등장 횟수를 표기
# 첫번째 데이터 불러오기
data = pd.read_excel('SNS데이터/SNS_8.xlsx', encoding='utf-8')
```

```python
data.head()
```

```python
# 필요없는 열 제거 (본문으로만 등장횟수를 지정할 것임)
data.drop(['DOCID', 'TITLE'], axis=1, inplace=True)
```

```python
#DATE를 Datetime으로 바꾸기
data['DATE']=pd.to_datetime(data['DATE'], format='%Y%m%d%H%M%S')
```

```python
#CONTENT에 결측값 제거하기
data=data.dropna(axis=0)
```

```python
data.shape
```

```python
#데이터가 많아 메모리 효율을 위해 분할하기
data_list=[]
k=0
for i in range(23):
    if i!=0:
        j=i*1000
        data_list.append(data.iloc[k:j])
        k=j
```

```python
#형태소 분리해서 list에 넣기 (1000 번째마다)
list_docs=[]
for i in range(len(data_list)):
    print(i*1000,'번째 실행 중입니다.')
    for j in range(len(data_list[i])):
            list_docs.append(get_splited(data_list[i].iloc[j,2]))
```

```python
# 미세먼지 단어 갯수 세기
list_count=[]
for doc in list_docs:
    list_count.append(tf(doc,'미세먼지'))
```

```python
data['Count']=list_count
```
